```python
from pymongo import MongoClient
import os

MDB_USERNAME = os.getenv('MDB_USERNAME')
MDB_PASSWORD = os.getenv('MDB_PASSWORD')
if MDB_PASSWORD is None or MDB_USERNAME is None:
    raise Exception('MDB_USERNAME and MDB_PASSWORD must be set in .env file')
mdb_client = MongoClient(
            host = "mongodb://localhost:27017/",
            username=MDB_USERNAME,
            password=MDB_PASSWORD,
        )

mdb_client.list_database_names()
db_news = mdb_client['news']
db_news.list_collection_names()
collection_articles = db_news['articles']
articles = collection_articles.find()
```

```python
def get_text(article):
    chunks = []
    # format_string += Fore.RED + article["url"] + Style.RESET_ALL + "\n"
    # format_string += Fore.GREEN + article["title"] + Style.RESET_ALL + "\n"

    for chunk in article["chunks"]:
        chunk_string = chunk["main"]
        for bullet in chunk["bullets"]:
            chunk_string += bullet
        chunk_string += "\n\n"
        chunks.append(chunk_string)

    return "".join(chunks)
```

```python
article = articles[1]
print(article["chunks"])
```

```python
article_texts = get_text(article)
print(article_texts)
```
